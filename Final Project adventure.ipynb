```python
dt['Name']=dt['Name'].str.title().str.strip().str.replace(r'[^a-zA-Z\s]', '', regex=True)
dt['CountryRegionCode']=dt['CountryRegionCode'].str.title().str.strip().str.replace(r'[^a-zA-Z\s]', '', regex=True)
dt['Group']=dt['Group'].str.title().str.strip().str.replace(r'[^a-zA-Z\s]', '', regex=True)
dt
```

```python
# Convert 'OrderDate' to datetime and extract year and month
dt['OrderDate']=pd.to_datetime(dt['OrderDate'])
dt['ShipDate']=pd.to_datetime(dt['ShipDate'])
dt['year']=dt['OrderDate'].dt.year
dt['month']=dt['OrderDate'].dt.month
dt['day']=dt['OrderDate'].dt.day
dt['shipday']=dt['ShipDate'].dt.day
dt
```

```python
# Extract quarter from the 'OrderDate'
dt['Quarter'] = dt['OrderDate'].dt.to_period('Q')
dt
```

```python
dt['Fulfillment Time']=dt['day']-dt['shipday']
dt
```

```python
#Total Sales by Territory
total_sales_by_territory = dt.groupby('Name')['TotalDue'].sum().reset_index()
plt.figure(figsize=(6,4))
plt.bar(total_sales_by_territory['Name'], total_sales_by_territory['TotalDue'])
plt.title('Total Sales by Territory')
plt.xlabel('Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
```

```python
#Sales Trend Over Time
sales_trend =dt.groupby('year')['TotalDue'].sum().reset_index()
plt.plot(sales_trend['year'], sales_trend['TotalDue'], marker='o', color='b')
plt.title('yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
```

```python
#top 10 customers by customerID
top_customers=dt.groupby('CustomerID')['TotalDue'].sum().nlargest(10)
top_customers.plot(kind='barh')
plt.title('Top 10 Customers by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Customer ID')
plt.xticks(rotation=45)
```

```python
#Sales Distribution
sales_distribution = dt.groupby('Name')['TotalDue'].sum()
plt.figure(figsize=(8, 8))
plt.pie(sales_distribution, labels=sales_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Territory')
```

```python
#Sales Trend Over quarter
sales_trend_over_quarter =dt.groupby('Quarter')['TotalDue'].sum()
sales_trend_over_quarter.plot(kind='bar', color='b', edgecolor='black')
plt.title('Sales Trend over Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
```

```python
#Sales Trend Over quarter (repeated)
sales_trend_quarterly =dt.groupby('Quarter')['TotalDue'].sum()
plt.figure(figsize=(10, 6))
plt.plot(sales_trend_quarterly.index.astype(str), sales_trend_quarterly, marker='o', linestyle='-', color='b')
plt.title('Sales Trend Over Quarters')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(rotation=45) 
```

```python
#Order Fulfillment Analysis
fulfillment_analysis =dt.groupby('Name')['Fulfillment Time'].mean()
fulfillment_analysis.plot(kind='bar')
plt.title('Average Order Fulfillment Time by Territory')
plt.xlabel('Territory Name')
plt.ylabel('Average Fulfillment Time (Days)')
```
